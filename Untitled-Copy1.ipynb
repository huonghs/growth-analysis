**Change period to date/time**

```python
octantis["Period"] = pd.to_datetime(octantis["Period"], format="%m/%d/%y")
octantis["Year"] = octantis["Period"].dt.year
octantis["Month"] = octantis["Period"].dt.month
octantis["Quarter"] = octantis["Period"].dt.quarter
octantis['Quarter_Year'] = octantis['Year'].astype(str) + '_' + octantis['Quarter'].astype(str)
octantis.head()
```

```python
octantis_sale = octantis[octantis["VariableName"] == "OCTANTIS RETAIL SALE UNITS"]
octantis_sale.head()
```

```python
octantis_sale.info()
```

```python
data1 = octantis_sale.groupby(['Year','Month']).sum()['VariableValue']
data1_1 = data1.unstack()
data1_1
```

```python
df5=octantis.pivot_table(index=['Period','Month','Quarter','Quarter_Year','Year'], columns='VariableName', values='VariableValue', aggfunc = 'sum')
```

```python
df5.reset_index(inplace=True)
```

```python
sale = df5.loc[:,['Period','Month','Quarter','Year','OCTANTIS RETAIL SALE UNITS']]
```

```python
sale
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(y = "OCTANTIS RETAIL SALE UNITS",x = "Month", hue = 'Year', data = sale, ci = None, markers = True, palette=['silver','darkorchid'])
plt.ylabel("Octantis T800 Sale Units")
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(y = "OCTANTIS RETAIL SALE UNITS",x = "Period", data = sale, color = "darkorchid")
plt.title('Sale by Period')
```

```python
sale.groupby(['Year']).sum()['OCTANTIS RETAIL SALE UNITS']
```

**Sale Units by Geography**

```python
df4=octantis.pivot_table(index=['Period','Month','Quarter','Quarter_Year','Year','Geography'], columns='VariableName', values='VariableValue', aggfunc = 'sum')
```

```python
df4.reset_index(inplace=True)
```

```python
sale_by_geo = df4.groupby(['Geography']).sum()['OCTANTIS RETAIL SALE UNITS']
sale_by_geo
```

```python
sale_by_geo = sale_by_geo.to_frame()
```

```python
sale_by_geo.reset_index(inplace=True)
```

```python
list(sale_by_geo)
```

```python
plt.figure(figsize=(7,8))
sns.barplot(x = "OCTANTIS RETAIL SALE UNITS", y = "Geography", data = sale_by_geo, color = "darkorchid")
plt.title('Sale by Location')
```
